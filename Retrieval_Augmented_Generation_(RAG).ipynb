```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from transformers import pipeline

# Load models out-of-the-box
print("Initializing core task models (this takes a moment)...")
classifier_node = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
generator_node = pipeline("text-generation", model="gpt2")
print("Models loaded successfully!")

# UI Components
text_input = widgets.Textarea(
    value="Defect found in the structural framework of the solar photovoltaic array. "
          "The maximum power point tracking unit failed to initialize during secondary testing in Nagpur.",
    description='Input Text:', layout=widgets.Layout(width='85%', height='80px')
)
task_selector = widgets.ToggleButtons(
    options=['Classification', 'Text Labelling', 'Text Generation', 'Summarisation'],
    description='Core Task:',
    button_style='info'
)
execute_btn = widgets.Button(description="Execute Task", button_style='success')
task_output = widgets.Output()

def process_task(b):
    with task_output:
        clear_output()
        raw_text = text_input.value
        task = task_selector.value

        print(f"Executing [{task}] on input text...\n" + "-"*50)

        if task == 'Classification':
            # Sorting into logical folders
            labels = ["Hardware Engineering", "Software Glitch", "Finance", "Music"]
            res = classifier_node(raw_text, candidate_labels=labels)
            print(f"Top Assigned Category: {res['labels'][0]} (Confidence: {res['scores'][0]*100:.1f}%)")
            print("\n=== Why this output arrived ===")
            print(f"The model uses self-attention to link your words like 'solar array' and 'photovoltaic' to the semantic concept of '{res['labels'][0]}'. It acts like an automated filing cabinet sorting system.")

        elif task == 'Text Labelling':
            # Identifying domain concepts inside a sentence
            labels = ["Technical Component", "Geographic Location", "Document Action"]
            res = classifier_node(raw_text, candidate_labels=labels)
            print("Detected Language Attributes:")
            for l, s in zip(res['labels'], res['scores']):
                print(f" - {l}: Mapping Strength {s*100:.1f}%")
            print("\n=== Why this output arrived ===")
            print("Instead of putting the whole text in one bucket, the model measures how closely the individual nouns tag against specific structural definitions. It acts as an automated tagger.")

        elif task == 'Text Generation':
            # Generating continuation sentences
            prompt = f"System Log Warning: {raw_text} Therefore, the engineer must"
            res = generator_node(prompt, max_new_tokens=25, num_return_sequences=1, pad_token_id=50256)
            print(f"Generated Completion:\n... {res[0]['generated_text'][len(prompt):]}")
            print("\n=== Why this output arrived ===")
            print("The model calculates the mathematical probability of words that naturally follow your sentence. It reads 'engineer must' and chooses logical professional actions based on its training patterns.")

        elif task == 'Summarisation':
            # Compressing information using explicit prompt triggers
            prompt = f"Text: {raw_text}\n\nSummary of the core issue:"
            res = generator_node(prompt, max_new_tokens=20, num_return_sequences=1, pad_token_id=50256)
            summary = res[0]['generated_text'][len(prompt):].strip()
            print(f"Extracted Core Summary:\n{summary}")
            print("\n=== Why this output arrived ===")
            print("By appending a clean prompt instruction line ('Summary of the core issue:'), we guide the transformer's attention map to look back at the nouns and verbs, pulling out only the densest informational keywords.")

execute_btn.on_click(process_task)
display(text_input, task_selector, execute_btn, task_output)
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load a vector embedding model out-of-the-box
print("Loading local vector embedding node...")
rag_embedder = SentenceTransformer('all-MiniLM-L6-v2')

# Simulating a LangChain/LlamaIndex document store
document_chunks = [
    "To extract structured tables from unstructured PDFs, write a Python script using python-docx.",
    "The on-road price of automatic compact SUVs like the Maruti Brezza in Nagpur varies by trim.",
    "A classical bamboo flute practice schedule requires strict finger dexterity exercises.",
    "A step-up systematic investment plan yields compound long-term capital wealth.",
    "A must visit place in Pune is FC Road"
]
doc_vectors = rag_embedder.encode(document_chunks)

# UI Components
user_query = widgets.Text(
    value="How can I automate document table extraction?",
    description='Query:', layout=widgets.Layout(width='70%')
)
search_btn = widgets.Button(description="Query Database", button_style='primary')
rag_viz_output = widgets.Output()

def run_rag_query(b):
    with rag_viz_output:
        clear_output()
        # Convert user question into the same mathematical vector space
        query_vector = rag_embedder.encode([user_query.value])

        # Calculate matching scores
        scores = cosine_similarity(query_vector, doc_vectors)[0]
        best_match_idx = np.argmax(scores)

        print(f"Step 1 (Framework Retrieval): Scanning vector store for: '{user_query.value}'")
        print(f"Top Chunk Retrieved:\n-> \"{document_chunks[best_match_idx]}\" (Match Score: {scores[best_match_idx]*100:.1f}%)\n")
        print("Step 2 (Augmentation Prompt created for the LLM):")
        print(f"\"\"\"Context: {document_chunks[best_match_idx]}\nQuestion: {user_query.value}\nAnswer based only on Context:\"\"\"\n")

        # Render the mathematical correlation chart
        plt.figure(figsize=(9, 3))
        colors = ['orange' if i == best_match_idx else 'lightgray' for i in range(len(document_chunks))]
        short_labels = [f"Chunk {i+1}: " + c[:30] + "..." for i, c in enumerate(document_chunks)]

        plt.barh(short_labels, scores, color=colors)
        plt.xlabel("Semantic Distance / Similarity Score")
        plt.title("RAG Framework Selection Analysis")
        plt.xlim(0, 1)
        plt.gca().invert_yaxis()
        plt.show()

        print("=== Why this output arrived ===")
        print("Frameworks like LangChain convert sentences into long lists of coordinates (embeddings).")
        print("Even if your question doesn't use the exact words as the text chunk, their spatial directions match closely in vector space. The chunk highlighted in orange is the context sent directly to the LLM so it can answer without hallucinating.")

search_btn.on_click(run_rag_query)
display(user_query, search_btn, rag_viz_output)
```
